# Neural Networks

A neural network is a kind of machine learning model made up of small units called neurons that are linked together in layers. It learns from data by adjusting weights and biases to find patterns and connect inputs with outputs. This allows it to perform tasks like classifying data, making predictions, and recognizing patterns. \
comsol_ref_definitions.19.050.08.png \
The main components of a neural network architecture are:

*   Input Layer
*   Output Layer
*   Neurons (Nodes)
*   Weights and Biases
*   Activation Functions
*   Loss Functions


A neural network learns to convert raw input data into meaningful information that helps it make accurate predictions. It does this through two key components: weights and biases. Weights determine how much influence each input has on a neurons output, while biases allow the network to adjust the output up or down, making it more flexible. During training, the network continuously updates these weights and biases by comparing its predictions with the true labels and calculating the error using a loss function. To reduce this error, the network applies an algorithm called backpropagation, which adjusts the parameters to improve performance.\
 Typically, the training process involves four main steps: forward propagation, error calculation using the loss function, backpropagation, and updating the parameters.These steps are repeated multiple times throughout the training data. With each iteration, the network gradually adjusts its internal parameters, improving its predictions and bringing them closer to the correct results. Over time, it learns an optimal set of weights and biases that minimize errors and perform effectively on new, unseen data. The main mechanism driving this learning process is backpropagation which enables neural networks with millions or even billions of parameters to discover meaningful patterns from large datasets.

## Loss Function

A loss function is a mathematical function used to measure model performance by calculating the deviation of a models predictions from the correct, ground truth predictions. Optimizing a model involves adjusting model parameters to minimize the output of some loss function.
